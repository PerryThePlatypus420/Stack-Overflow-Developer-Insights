```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('/kaggle/input/stack-overflow-2023-developers-survey/survey_results_public.csv')
df.head()
```

```python
print("(rows, columns): ", df.shape)
```

```python
print("Column Names of the dataset:\n\n", df.columns)
```

```python
df = df.rename({
    'ConvertedCompYearly':'Salary'
},axis=1)

print("Column Names of the dataset:\n\n", df.columns)
```

```python
# Identify indices of rows with null values in the "Salary" column
null_salary_indices = df[df['Salary'].isnull()].index

# print(null_salary_indices)

# Keep the first 30,000 null rows, remove the rest
indices_to_remove = null_salary_indices[12000:]
df = df.drop(indices_to_remove)

# Verify the shape of the DataFrame after removal
print("(rows, columns): ", df.shape)

df.head()
```

```python
# Replacing the null currency cells with standard 'USD' currency
standard_curr = 'USD	United States dollar'
df.fillna({'Currency': standard_curr}, inplace=True)

unique_curr = df['Currency'].unique()
print("Unique currency names in the dataset:\n\n", unique_curr)
```

```python
# Calculate mean salary for each currency category
mean_salary_per_currency = df.groupby('Currency')['Salary'].mean()
# print(mean_salary_per_currency)

# Replace NaN values in 'Salary' column with mean salary for each currency category
for currency, mean_salary in mean_salary_per_currency.items():
    df.loc[df['Currency'] == currency, 'Salary'] = df.loc[df['Currency'] == currency, 'Salary'].fillna(mean_salary)
    
```

```python
# Finding the number of null values in 'Currency'
null_values_curr = df['Currency'].isnull().sum()
print("Number of null values in 'Currency':", null_values_curr)


# Finding the number of null values in 'Salary'
null_values_sal = df['Salary'].isnull().sum()
print("Number of null values in 'Salary':", null_values_sal)
```

```python
# Printing 'Currency' and 'Salary' columns where 'Salary' column has null values
rows_with_null_salary = df[df['Salary'].isnull()][['Currency', 'Salary']]

print("Currency and Salary columns where 'Salary' column has null values:")
print(rows_with_null_salary)
```

```python
# Dropping these rows with null values in the 'Salary' column
df.dropna(subset=['Salary'], inplace=True)

# Deleting the first row of the DataFrame
df.drop(df.index[0], inplace=True)

print("(rows, columns): ", df.shape)
```

```python
# Keeping only the useful and important columns
df = df[['Age','Employment','RemoteWork','EdLevel','YearsCode', 'DevType',
         'OrgSize', 'Country', 'Currency', 'LanguageHaveWorkedWith',
         'LanguageWantToWorkWith', 'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith', 
         'PlatformHaveWorkedWith', 'PlatformWantToWorkWith', 'WebframeHaveWorkedWith',
         'WebframeWantToWorkWith', 'OpSysProfessional use', 
         'WorkExp','Industry','Salary'
        ]]

print("The cleaned Dataset is as follows:\n", df.columns)

# df.describe()
print(df.dtypes)
```

```python
mean_work_exp = int(round(df['WorkExp'].mean()))
df.fillna({'WorkExp': mean_work_exp}, inplace=True)

wrkexp_null_counts = df['WorkExp'].isnull().sum()
print("Total nulls in 'WorkExp': ", wrkexp_null_counts)
```

```python
# Fill missing values in the entire DataFrame with an empty string
df.replace(to_replace=pd.NA, value='', inplace=True)

# Replace 'NaN' values with an empty string
df.replace(to_replace=np.nan, value='', inplace=True)

print("All the null values replaced with an empty string, ''.")
```

```python
# Count the number of nulls in each column of the DataFrame
null_counts = df.isnull().sum()

# Total number of nulls in the DataFrame
total_nulls = null_counts.sum()

print("Total number of nulls in the DataFrame:", total_nulls)
```
